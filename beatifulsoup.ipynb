```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import time

def scrape_naver_reviews(url, max_clicks=3):
    # 1. 크롬 드라이버 설정
    chrome_options = Options()
    chrome_options.add_experimental_option("detach", True) # 브라우저 꺼짐 방지
    # chrome_options.add_argument("--headless") # 백그라운드 실행을 원하면 주석 해제

    driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=chrome_options)
    driver.get(url)
    driver.implicitly_wait(10) # 페이지 로딩 대기

    print("페이지 로딩 완료. '더보기' 버튼 클릭을 시작합니다.")

    # 2. '더보기' 버튼 클릭 루프 (원하는 만큼 리뷰를 불러오기 위해)
    for i in range(max_clicks):
        try:
            # 주의: 네이버는 클래스명이 주기적으로 바뀝니다.
            # 현재 '더보기' 버튼의 태그나 클래스를 개발자 도구(F12)로 확인 후 업데이트해야 할 수 있습니다.
            # 보통 <a> 태그나 버튼 태그 안에 '더보기' 텍스트가 있습니다.
            more_button = WebDriverWait(driver, 5).until(
                EC.element_to_be_clickable((By.XPATH, "//a[contains(text(), '더보기')] | //span[text()='더보기']/parent::a"))
            )
            more_button.click()
            time.sleep(1.5) # 클릭 후 로딩 대기 (너무 짧으면 IP 차단 위험)
            print(f"더보기 버튼 클릭 완료 ({i+1}/{max_clicks})")
        except Exception as e:
            print("더보기 버튼이 더 이상 없거나 클릭할 수 없습니다.")
            break

    # 3. 로딩된 페이지의 HTML 파싱
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')

    # 4. 리뷰 데이터 추출
    # 주의: 리뷰 내용이 담긴 클래스명(예: 'zPfVt')은 네이버에서 랜덤으로 생성하여 자주 바뀝니다.
    # 반드시 크롬 개발자 도구(F12)를 열어 실제 리뷰 텍스트가 담긴 태그의 클래스명을 확인하고 변경하세요.
    review_elements = soup.find_all('span', class_='pui__vn15t2') # <--- 여기를 수정해야 할 확률이 높습니다!

    reviews = []
    for index, element in enumerate(review_elements):
        review_text = element.get_text(strip=True)
        if review_text:
            reviews.append(review_text)
            print(f"[{index + 1}] {review_text}")

    # 5. 브라우저 종료
    driver.quit()

    return reviews

# 실행 예시 (강남역 스타벅스 등 특정 장소의 방문자 리뷰 모바일 링크)
# 주소 형식: https://m.place.naver.com/restaurant/[식당고유번호]/review/visitor
target_url = "https://m.place.naver.com/restaurant/36181964/review/visitor"
extracted_reviews = scrape_naver_reviews(target_url, max_clicks=5)

print(f"총 {len(extracted_reviews)}개의 리뷰를 수집했습니다.")
```
